```python
# classes
class RoadUser:
    def __init__(self, position, velocity):
        self.name = ''
        self.position = position        
        self.velocity = velocity

    def move(self, deltaT):
        self.position += deltaT*self.velocity

    def getType(self):
        return 'User'

    def __str__(self):
        return '{} @ {}'.format(self.getType(), self.position)

    def draw(self):
        pass

class PassengerVehicle(RoadUser):
    def getType(self):
        return 'Véhicule'

class Pedestrian(RoadUser):
    def getType(self):
        return 'Piéton'

class Cyclist(RoadUser):
    def getType(self):
        return 'Cycliste'

roadUsers = [PassengerVehicle(1, 5.),
             Pedestrian(10, 1.),
             Cyclist(15, 2.)]

for r in roadUsers:
    print(r)
```

```python
# algorithme factorielle
def Factorielle(N):
    M=1
    F=1
    while M != N:
        M=M+1
        F=F*M
        print("F={},M={}".format(F, M))
    return F

N = int(input('Entrer un entier positif: '))
print('Résultat de factorielle de {}: {}'.format(N, Factorielle(N)))
```
